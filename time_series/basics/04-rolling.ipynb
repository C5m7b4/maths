```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds = pd.read_csv('../../data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
ds['Close'].plot(figsize=(12,5))
```

```python
ds['Close'].plot(figsize=(12,5))
ds.rolling(window=30).mean()['Close'].plot()

plt.show()
```

```python
ds['Close:30 day mean'] = ds['Close'].rolling(window=30).mean()
```

```python
ds[['Close', 'Close:30 day mean']].plot(figsize=(12,5))
```

```python
ds['Close'].expanding().mean().plot(figsize=(12,5))
```
