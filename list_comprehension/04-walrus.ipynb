```python
from urllib.parse import urlparse
from typing import Union
```

```python
def return_scheme(st: str) -> str | bool:
    """Return scheme if valid URL, else return false"""
    url = urlparse(st)
    if all([url.scheme, url.netloc]):
        return url.scheme
    return False
    
```

```python
print(return_scheme('https://www.google.com'))
```
