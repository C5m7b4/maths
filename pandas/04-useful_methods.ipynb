## apply  using multiple columns

```python
ds.head()
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'generous'
    else:
        return 'other'
```

```python
ds[['total_bill', 'tip']].apply(lambda x: quality(x['total_bill'], x['tip']), axis=1)
```

```python
ds['quality'] = np.vectorize(quality)(ds['total_bill'], ds['tip'])
```

```python
ds.head()
```

## time vectorize function

```python
import timeit
```

```python
setup = '''
import numpy as np
import pandas as pd
ds = pd.read_csv('../data/tips.csv')
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'generous'
    else:
        return 'other'
'''
```

```python
stmt_one = '''ds[['total_bill', 'tip']].apply(lambda x: quality(x['total_bill'], x['tip']), axis=1)'''
stmt_two = '''ds['quality'] = np.vectorize(quality)(ds['total_bill'], ds['tip'])'''
```

```python
timeit.timeit(setup=setup, stmt=stmt_one, number=1000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_two, number=1000)
```
