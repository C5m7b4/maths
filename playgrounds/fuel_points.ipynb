```python
import numpy as np
import pandas as pd
```

```python
accrued = pd.read_csv('../data/Jason_points_461_from_jan.csv', dtype={'customer_number': str})
```

```python
accrued['customer_number'] = accrued['customer_number'].astype(str)
```

```python
accrued.head()
```

```python
redeemed = pd.read_csv('../data/Json_451_redemptions.csv', dtype={'customer_number': str})
```

```python
redeemed['customer_number'] = redeemed['customer_number'].astype(str)
```

```python
redeemed.head()
```

```python
common_customers = set(accrued['customer_number']) & set(redeemed['customer_number'])
```

```python
common_customers
```

```python
accrued_common = accrued[accrued['customer_number'].isin(common_customers)]
redeemed_common = redeemed[redeemed['customer_number'].isin(common_customers)]
```

```python
merged = pd.merge(
    accrued_common,
    redeemed_common,
    on='customer_number',
    suffixes=('_accrued', '_redeemed'),
)
```

```python
grouped = merged.groupby('customer_number')
```

```python
print(grouped.get_group(list(common_customers)[0]))
```

```python
for cust, df in grouped:
    df.to_csv(f'customers/customer_{cust}_activity.csv', index=False)
```

# V2

```python
accrued_summary = accrued.groupby('customer_number')['points'].sum().reset_index()
accrued_summary.rename(columns={'points': 'total_accrued_points'}, inplace=True)
```

```python
redeemed_summary = redeemed.groupby("customer_number")['points'].sum().reset_index()
redeemed_summary.rename(columns={'points': 'total_redeemed_points'}, inplace=True)
```

```python
merged_summary = pd.merge(
    accrued_summary,
    redeemed_summary,
    on='customer_number',
    how='inner'
)
```

```python
merged_summary.to_csv('customers/customer_points_summary.csv', index=False)
```
