## CoVariance

$
\Large
\text{CoVar}=\frac{\sum_{i}{(x_i-\bar{x}) \cdot (y_i - \bar{y})}}{N-1}
$

## Correlation

$
\Large
\text{Correlation} = \frac{ \sum_{i=1}^{N} (x_i-\bar{x}) \cdot (y_i - \bar{y}) }{ (N-1)\,\sigma_{x}\,\sigma_{y} }
$

here we divide by (N-1) times the standard deviation of x times the standard deviation of y

```python
hills.cov()
```

```python
hills.var()
```

```python
def covariance(ds, x, y):
    return ((ds[x] - ds[x].mean()) * (ds[y] - ds[y].mean())).sum()/ (len(ds) - 1)
```

```python
cov_dist_climb = covariance(hills, 'dist', 'climb')
cov_dist_climb
```

```python
hills.corr()
```

```python
def correlation(ds, x, y):
    return (((ds[x] - ds[x].mean()) * (ds[y] - ds[y].mean())).sum()/(len(ds) - 1)) / (ds[x].std() * ds[y].std())
```

```python
correlation(hills, 'dist', 'time')
```
