```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

$
t = f'_a(x - x_a) + f_a
$

```python
x = sym.symbols('x')
```

```python
f = x**2
df = sym.diff(x)

xa = 1

fa = f.subs(x, xa)
dfa = df.subs(x, xa)
dfa
```

```python
xx = np.linspace(-2, 2, 200)
f_fun = sym.lambdify(x, f)(xx)
df_fun = sym.lambdify(x, df)(xx)


# compute the tangent line
tanline = dfa * (xx-xa) + fa

plt.plot(xx, f_fun, label='f(x)')
plt.plot(xx, tanline, label='tangent')


plt.plot(xa, fa, 'ro')
plt.legend()
plt.grid()
plt.axis('square')
plt.axis([-2, 2, -.5, 2])

ax = plt.gca()
plt.plot(ax.get_xlim(), [0,0], 'k--')
plt.plot([0, 0], ax.get_ylim(),'k--')
plt.show()
```
