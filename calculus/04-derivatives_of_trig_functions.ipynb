```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

## Circular derivatives

$
\Large
\frac{d}{dx} = cos(x) = -sin(x)
$

$
\Large
\frac{d}{dx} = -sin(x) = -cos(x)
$

$
\Large
\frac{d}{dx} = -cos(x) = sin(x)
$

$
\Large
\frac{d}{dx} = sin(x) = cos(x)
$

```python
q = sym.symbols('q')
```

```python
sym.diff(sym.cos(q))
```

```python
sym.diff(sym.sin(q))
```

```python
f = sym.cos(q)


for i in range(0, 8):
    display( Math( '\\frac{d}{dx}%s = %s' %(sym.latex(f), sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
import sympy.plotting.plot as symplot
```

```python
f = sym.cos(q)

for i in range(0, 4):
    if i == 0:
        p = symplot(f, show=False, label=r'$%s$' %sym.latex(f))
    else:
        p.extend(symplot(f, show=False, label=r'$%s$' %sym.latex(f)))
    f = sym.diff(f)

p.legend=True
p.xlim=[-3, 3]
p.show()
```

## excercise

```python
x, a = sym.symbols('x, a')

f = sym.cos(x + sym.sin(x)) + a

for ai in range(0, 4):
    if ai == 0:
        p = symplot(f.subs(a, ai), show=False, label=r'$%s$' %sym.latex(f.subs(a, ai)))
    else:
        p.extend(symplot(f.subs(a, ai), show=False, label=r'$%s$' %sym.latex(f.subs(a, ai))))

p.title = 'The Functions'
p.legend=True
p.show()
```

```python
for ai in range(0, 4):
    if ai == 0:
        p = symplot(sym.diff(f.subs(a, ai)), show=False, label=r'$%s$' %sym.latex(f.subs(a, ai)))
    else:
        p.extend(symplot(sym.diff(f.subs(a, ai)), show=False, label=r'$%s$' %sym.latex(f.subs(a, ai))))

p.title = 'Their derivatives Functions'
p.legend=True
p.show()
```
