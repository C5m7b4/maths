```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
```

```python
fx = x**3

lim_point = 1.5
lim = sym.limit(fx, x, lim_point)
display( Math( '\\lim_{x \\to %g} %s = %g' %(lim_point, sym.latex(fx), lim)))
```

```python
1.5**3
```

```python
fxx = sym.lambdify(x, fx)
xx = np.linspace(-5, 5, 200)

plt.plot(xx, fxx(xx))
plt.plot(lim_point, lim, 'ro')
plt.show()
```

```python
fx = (x**2)/(x-2)

fxx = sym.lambdify(x, fx)
xx = np.linspace(1, 3, 102)

lim_point = 2
lim = sym.limit(fx, x, lim_point, dir='-')


plt.plot(xx, fxx(xx))

plt.show()

display( Math( '\\lim_{x \\to %g^-} %s = %g' %(lim_point, sym.latex(fx), lim)))
```

```python
x = sym.symbols('x')

fx = sym.sqrt(x+1)*sym.exp(-x)
gx = sym.cos(x+sym.sin(x))

fxx = sym.lambdify(x, fx)
gxx = sym.lambdify(x, gx)

xx = np.linspace(0, 10, 100)

plt.plot(xx, fxx(xx), label='f(x)')
plt.plot(xx, gxx(xx), label='g(x)')
plt.legend()
plt.show()

```

```python
# compute the limits of f and g
lim_point = 5

lim_fx = sym.limi(fx, x, lim_point)
lim_gx = sym.limit(gx, x, lim_point)

```
