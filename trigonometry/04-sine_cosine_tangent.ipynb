```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = np.linspace(0, 6*np.pi, 100)

plt.plot(x, np.sin(x), label='$\\text{sin}$($\\theta$)')
plt.plot(x, np.cos(x), label='cos($\\theta$)')
plt.legend()
plt.xlabel('Radians')
plt.ylabel('fct value')
plt.show()
```

```python
th = np.linspace(0, 4*np.pi, 300)

plt.plot(th, np.tan(th), 'k')


plt.show()
```

```python
ang = np.random.rand() * 2 * np.pi
tan = np.tan(ang)

sc = np.sin(ang) / np.cos(ang)

tan - sc
```

```python
# cos**2 + sin**2 = 1
thetas = np.linspace(0, 2*np.pi, 13)
np.cos(thetas)**2 + np.sin(thetas)**2
```

## excersice

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)

y1 = np.sin(x + np.cos(x))
y2 = np.cos(x + np.sin(x))

plt.plot(x, y1, label='sin(x+cos(x))')
plt.plot(x, y2, label='cos(x + sin(x))')

plt.legend()
plt.show()
```

```python
thsym = sym.pi*7/6
theta = float(thsym)

#plot the unit circle
x = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(x), np.sin(x), 'k')

plt.plot([-1, 1], [0, 0], 'k--')
plt.plot([0, 0], [-1, 1], 'k--')

# plot the arc
plt.plot(np.cos(np.linspace(0, theta)), np.sin(np.linspace(0, theta)), 'p', linewidth=3)
plt.plot([0, np.cos(theta)], [0, np.sin(theta)], color='purple')
plt.plot(np.cos(theta), np.sin(theta), 'bo')

plt.plot([0, 0], [0, np.sin(theta)], color='red')
plt.plot([0, np.cos(theta)], [np.sin(theta), np.sin(theta)], 'r:')

plt.plot([0, np.cos(theta)], [0, 0], 'b')
plt.plot([np.cos(theta), np.cos(theta)], [0, np.sin(theta)], 'b:')

plt.axis('square')
plt.title('$\\theta = %s$' %sym.latex(thsym))
plt.show()
```
