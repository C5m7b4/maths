```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
k = np.pi/6
m = 2.3

eul = m*np.exp(1j*k)
cis = m*( np.cos(k) + 1j*np.sin(k) ) # cos isin

cis, eul
```

```python
mag = np.abs(eul)
ang = np.angle(eul)

plt.polar([0, ang], [0, mag], 'r')
plt.polar(k, m,'bo')

plt.show()
```
