```python
import numpy as np
import matplotlib.pyplot as plt
```

## random distribution

```python
nums = np.random.rand(1000)

minval = 2
maxval = 17

nums = nums * (maxval-minval) +  minval

plt.plot(nums, 's')
plt.show()


plt.hist(nums)
plt.show()
```

## normal distribution

```python
nums = np.random.randn(1000) # normal distribution

plt.plot(nums, 's', alpha=0.5)
plt.show()


plt.hist(nums)
plt.show()
```

## excercise

```python
desired_mean = 15
desired_std = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums)
nums = nums / np.std(nums)

nums = nums * desired_std + desired_mean
#* desired_std + desired_mean

plt.subplot(121)
plt.plot(nums, 's', alpha=0.5)

plt.subplot(122)
plt.hist(nums)

plt.show()

print('This distribution has a mean of %s and a std of %s' %(np.mean(nums), np.std(nums)))
```

## random phase angles

```python
angs = np.random.randn(100) # normal distribution

for i in angs:
    plt.plot([0, np.cos(i)], [0, np.sin(i)])

plt.axis('square')
plt.axis([-1, 1, -1, 1])
plt.axis('off')
plt.show()    
```

```python
angs = np.random.rand(100) # uniformally distributed

for i in angs:
    plt.plot([0, np.cos(i)], [0, np.sin(i)])

plt.axis('square')
plt.axis([-1, 1, -1, 1])
plt.axis('off')
plt.show() 
```

```python
angs = np.random.rand(100) * 2 * np.pi

for i in angs:
    plt.plot([0, np.cos(i)], [0, np.sin(i)]) # these are in terms of radians

plt.axis('square')
plt.axis([-1, 1, -1, 1])
plt.axis('off')
plt.show() 
```
