```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

$$
\Large
z^n = 1, \quad z \quad \in \quad \mathbb{C}
$$

$$
\Large
z = e^{2 \pi ik/n}
$$

$$
\Large
k = 0, 1, ...n-1

```python
# we could use 1j as the imaginary operator which is the same as ik in some notation
1j
```

```python
# or we could use sym.I
sym.I
```

```python
n = 4

for k in range(0, n):
    z = sym.exp(2*sym.pi*sym.I*k/n)
    display( Math(  '(%s)^{%s} \\Rightarrow %s' %(sym.latex(z), n, sym.latex(z**n))))
```

```python
n = 14

for k in range(0, n):
    z = np.exp(2*np.pi*1j*k/n)
    plt.plot([0, np.real(z)], [0, np.imag(z)])

plt.axis('square')
x = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(x), np.sin(x),color='gray')
plt.show()   
```

## excercise

this is our old formula

$ \Large 2 \pi ik/n $

lets try it like this:

$ \Large \frac{k2 \pi i}{n} $

$
\Large
ke^{\frac{k2 \pi i}{n}}
$

```python
n = 2000

color = np.linspace(0, .9, n)

for k in range(0, n):
    z = k * np.exp(2*np.pi*1j*k/n)
    plt.plot([0, np.real(z)], [0, np.imag(z)], linewidth=2, color=[color[k], color[k], color[k]])

plt.axis('square')
plt.axis('off')
plt.show()  
```
