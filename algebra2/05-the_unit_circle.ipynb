```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(np.cos(x), np.sin(x), 'k')
plt.plot([-1, 1], [0, 0], '--', color=[.8, .8, .8], lw=0.5)
plt.plot([0, 0],[-1, 1], 'k',  lw=0.5)

angle = np.pi/4
plt.plot([0, np.cos(angle)], [0, np.sin(angle)])


plt.axis('square')
plt.xlabel('cos(x)')
plt.ylabel('sin(x)')


plt.show()
```

```python
x = np.linspace(-3, 1.83, 50)
k = np.exp(x)

for i in k:
    plt.plot([0, np.cos(i)], [0, np.sin(i)])


plt.axis('square')
plt.axis('off')
plt.show()    
```
