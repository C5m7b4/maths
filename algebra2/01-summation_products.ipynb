```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst = [1, 3, 4, 1, 6]
print(lst)
```

```python
np.sum(lst)
```

```python
np.prod(lst)
```

```python
1 * 3 * 4 * 1 * 6
```

```python
np.cumsum(lst)
```

```python
plt.plot(lst, 'rs-')
plt.plot(np.cumsum(lst), 'bs-')

plt.legend(['set', 'cumsum'])
plt.show()
```

```python
# summation

A = np.arange(1, 6)

suma = np.sum(A)
suma2 = np.sum(A**2)

ans1 = suma / suma2
ans2 = 1 / suma

print('option1 = %s  \noption2 = %s' %(ans1, ans2))
```

```python
# products

proda = np.prod(A)
prod2 = np.prod(A**2)

ans1 = proda / prod2
ans2 = 1 / proda

print('option1 = %s  \noption2 = %s' %(ans1, ans2))

```
