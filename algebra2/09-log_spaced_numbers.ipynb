```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.linspace(1, 2, 10)
```

```python
a = np.log10(1)
b = np.log10(2)
np.logspace(a, b, 10)
```

```python
a = 2
b = 100
n = 50
```

```python
li = np.linspace(a, b, n)
lo = np.logspace(np.log10(a), np.log10(b), n)


plt.plot(li, li, label='linear')
plt.plot(li, lo, label='log')
plt.legend()
plt.axis('square')
plt.show()
```
