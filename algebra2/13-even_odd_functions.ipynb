```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 20)

f_even = x**2
f_even_neg = (-x)**2

plt.plot(x, f_even)
plt.plot(x, f_even_neg, 'rd')

plt.show()
```

```python
f_odd =x**3

f_odd_neg = (-x)**3
f_neg_odd = -f_odd

plt.plot(x, f_odd)
plt.plot(x, f_odd_neg, 'rd')
plt.plot(x, f_neg_odd, 'g')

plt.show()

```

```python
theta = np.linspace(-2*np.pi, 2*np.pi, 39)

cos_fn   = np.cos(theta)
cos_fn_n = np.cos(-theta)

plt.plot(theta, cos_fn)
plt.plot(theta, cos_fn_n, 'rd')
plt.title("cosine(x)")
plt.show()

sin_fn   = np.sin(theta)
sin_fn_n = np.sin(-theta)

plt.plot(theta, sin_fn)
plt.plot(theta, sin_fn_n, 'rd')
plt.title("sin(x)")
plt.show()
```
