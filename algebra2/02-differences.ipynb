```python
x = np.linspace(-2, 2, 150)

f = x**2
df = np.diff(f)
dx= np.diff(x)

plt.plot(x, f, label='f')
plt.plot(x[:-1], df/dx, label='df/dx')
plt.legend()
plt.axis([-2,2, -1,1])
plt.grid()
plt.show()
```
