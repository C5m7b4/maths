```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
\Large
y = e^{\frac{-4ln(2)t^2}{h^2}}
$$


```python
t = np.linspace(-3, 3, 1001)
h = .999
```

```python
g = np.exp( -4 * np.log(2) * t**2 / h**2)
```

```python
plt.plot(t, g)

plt.xlabel('t')
plt.ylabel('y')
plt.title('Gaussian with FWHM=' + str(h)) # full width at half maximum
plt.show()
```

```python
N = 100

G = np.zeros((N, len(t)))
h = np.zeros(N)
```

```python
for i in range(N):
    h[i] = (i + 1) / N
    G[i,:] = np.exp( -4 * np.log(2) * t**2 / h[i]**2)

plt.pcolormesh(t,h,G)
plt.xlabel('t')
plt.ylabel('h')
plt.show()    
```
