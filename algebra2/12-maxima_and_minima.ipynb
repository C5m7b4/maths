```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst = [-4, 5.5, 1, 2]

maxval = np.max(lst)
maxvalIdx = np.argmax(lst)

lst[maxvalIdx]
```

```python
x = np.linspace(0, 2*np.pi, 20)

fx = -(np.cos(x) + np.sqrt(x))

fmax = np.max(fx)
fmaxIdx = np.argmax(fx)
print(fmax, fmaxIdx)

plt.plot(x, fx, 'bo-')
plt.plot(x[fmaxIdx], fmax, 'rs')

plt.show()
```

```python
x = np.linspace(-2, 2, 51)

fx = x**3 + x**4
minval = np.min(fx)
minIdx = np.argmin(fx)

plt.plot(x, fx, 'bo-')
plt.plot(x[minIdx], minval, 'rs')
plt.xlim([-1.2, .5])
plt.ylim([-.2, .2])
plt.show()
```
