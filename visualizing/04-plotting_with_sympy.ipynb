```python
import sympy as sym
import matplotlib.pyplot as plt
import sympy.plotting.plot as symplot
```

```python
x = sym.symbols('x')
```

```python
y = x**2
```

```python
p = symplot(y, show=False)
p.xlim = [0, 50]
p[0].line_color = 'm'
p.title = 'this is a nice plot'
print(p)
p.show()
```

```python
x, a = sym.symbols('x, a')
```

```python
expr = a/x

n = 1

p = symplot(expr.subs(a, n), (x, -5, 5), show=False)
p[0].label = '$y = %s$' %(sym.latex(expr.subs(a, n)))

p2 = symplot(expr.subs(a, 2), show=False)
p.extend( p2)
p[1].label = '$y = %s$' %(sym.latex(expr.subs(a, 2)))

p[1].line_color='r'
p.ylim = [-5, 5]
p.legend = True
p.show()

```
