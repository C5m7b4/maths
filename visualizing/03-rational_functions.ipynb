```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = range(-3, 4)

y = np.zeros(len(x))

for i in range(0, len(x)):
    y[i] = 2 - x[i]**2
plt.plot(x, y, 's-')
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = 2 + np.sqrt(abs(x))

plt.plot(x, y, 'm-')
plt.show()
```
