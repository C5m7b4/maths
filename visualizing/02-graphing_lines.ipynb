## slope intercept form

```python
#y = mx + b
```

```python
x = [-5, 5]
m = 2
b = 1
```

```python
y = [3, 4]

for i in range(0, len(x)):
    y[i] = m * x[i] + b

print(x,y)
```

```python
x = [-5, 5]
m = -.5
b = .1

y = [3, 4]

# for i in range(0, len(x)):
#     y[i] = m * x[i] + b

y = m * np.array(x) + b

plt.plot(x, y, label='y = %sx + %s' %(m, b))

plt.axis('square')
plt.xlim(x)
plt.ylim(x)
plt.grid()

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')
plt.legend()
plt.show()
```

```python
x = [-5, 5]
m = [.7, -1.25]
b = [-2, .75]

for i in range(0, len(m)):
    y = m[i] * np.array(x) + b[i]
    plt.plot(x, y, label='y = %sx + %s' %(m[i], b[i]) )

plt.axis('square')
plt.xlim(x)
plt.ylim(x)
plt.grid()

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.show()    
```
