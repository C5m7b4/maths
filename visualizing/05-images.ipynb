```python
A = np.zeros([10, 14])

for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        A[i,j] = 3 * i - 4 * j

plt.imshow(A)
#plt.plot([0, 3], [8, 2], 'r', linewidth=2)

for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        plt.text(j, i, int(A[i, j]), horizontalalignment='center', verticalalignment='center', color='black')

plt.set_cmap('Purples')
plt.show()        
```

```python
n  = 10
C = np.zeros([n, n])

for i in range(0, n):
    for j in range(0, n):
        C[i, j] = (-1)**(i+j) # pemdas

plt.imshow(C)        
plt.tick_params(labelleft=False, labelbottom=False)
# plt.savefig('C.png')
plt.show()
print(C)
```
