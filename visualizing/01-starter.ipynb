```python
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = 3
y = 5
```

```python
plt.plot(x, y, 'ro')
plt.axis('square')
plt.axis([-6,6, -6, 6])
plt.grid()
plt.show()
```

```python
x = [-4, 5, 2, 6, 3 -3]
y = [5, 2, 8, 4, -2, 5]
```

```python
for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label='point %s' %(i+1))


plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4, 2, 'r+')
axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)
axis.set_ylim(([ylim[0], 5.44]))
plt.show()
```

```python
x = sym.symbols('x')
```

```python
y = x**2 - 3*x
```

```python
for pnti in range(-11, 11):
    plt.plot(pnti, y.subs(x, pnti), 'o', label='point %s' %(pnti+1))

plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)
plt.show()    
```
